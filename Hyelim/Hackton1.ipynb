```python
def split_date_time(df):
    if 'visit_datetime' in df.columns:
        df.loc[:,'visit_datetime'] = pd.to_datetime(df['visit_datetime'])
        df.loc[:,'visit_date']  = [str(d.date()) for d in df['visit_datetime']]       
        df.loc[:,'visit_time']  = [str(d.time()) for d in df['visit_datetime']]       
    if 'reserve_datetime' in df.columns:
        df.loc[:,'reserve_datetime'] = pd.to_datetime(df['reserve_datetime'])
        df.loc[:,'reserve_date']  = [str(d.date()) for d in df['reserve_datetime']]       
        df.loc[:,'reserve_time']  = [str(d.time()) for d in df['reserve_datetime']]
        
    if 'visit_date' in df.columns:
        result = pd.merge(df, data['date_info'], how='inner', left_on='visit_date', right_on='calendar_date')
    if 'reserve_date' in df.columns:
        result = pd.merge(result,data['date_info'],how='inner', left_on='reserve_date', right_on='calendar_date')
    return result
```

```python
split_date_time(data['hpg_reserve'])
```
